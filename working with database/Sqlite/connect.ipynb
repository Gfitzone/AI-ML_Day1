```python
#connecting to the existing db using SQLite and if db doesnot exist, 
# then it will be created and finally a db object will be returned
import sqlite3 #it is a python database driver
conn=sqlite3.connect('garimaa.db')
print("database connected successfully !")
```

```python
#creating table in the above database
conn.execute('''CREATE TABLE  IF NOT EXISTS COMPANY
             (
             ID INT PRIMARY KEY NOT NULL,
             NAME TEXT NOT NULL,
             AGE INT NOR NULL,
             ADDRESS VARCHAR(100) NOT NULL,
             SALARY REAL);'''
             )
print("Table created successfully")
conn.close()
```

```python
import sqlite3
conn=sqlite3.connect('garimaa.db')

#inserting record in sqlite database table
conn.execute("INSERT INTO COMPANY (ID, NAME, AGE, ADDRESS, SALARY) VALUES(?,?,?,?,?)",(1,"GARIMA",21,"Wellington",110000.90))
   #VALUES(?,?,?,?) IS A PLACEHOLDER TO AVOID SYNTAX ERROR AND SQL INJECTION          
conn.execute("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY) VALUES(?,?,?,?,?)",(2,"DHANISTA",23,"New-Zealand",100000.34))
conn.execute("INSERT INTO COMPANY (ID,NAME,AGE,ADDRESS,SALARY) VALUES(?,?,?,?,?)",(3,"BHARANI",21,"Otago",120000.20))

conn.commit()
print("Record stored successfully !")
conn.close()

            
```

```python
#to fetch and display the records

conn=sqlite3.connect('garimaa.db')
print("opened database successfully")
print("Records:\n")
cursor=conn.execute("select * from COMPANY")
#for row in cursor:
 #   print(row)

#for more cleaner output we can also do
for x in cursor:
    print('ID:',x[0])
    print('NAME:',x[1])
    print('AGE:',x[2])
    print('ADDRESS:',x[3])
    print('SALARY:',x[4],"\n")
conn.close()
```

```python
#updating the records

conn=sqlite3.connect('garimaa.db')
conn.execute("UPDATE COMPANY SET SALARY=200000 WHERE ID=1")
conn.commit()
print("Total number of row updated:",conn.total_changes)

print("records\n")
cursor=conn.execute("select id, name, age, address,salary from company WHERE ID=1")
#yo sqlite case sensitive hudaina hai matlab yeha agi maile company, name age address etc sabai capital
#letter ma create gareko theye tara aaile update garda small letter ma lekhda pani yesle mazzlae operation garcha hai :)
for row in cursor:
    print("ID:",row[0])
    print("Name:",row[1])
    print("Age:",row[2])
    print("Address:",row[3])
    print("Salary:",row[4], "\n")
print("\n operation done successfully !")
conn.close()

```
