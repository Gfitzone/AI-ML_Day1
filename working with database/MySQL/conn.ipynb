```python
#to insert multiple rows in a table we use executemany*
import mysql.connector as sqlc
mydb=sqlc.connect(
    host="localhost",
    user="root",
    password="",
    database="garimaDB"
)
mycursor=mydb.cursor()
sql="""insert into tlb_students (id,NAME,AGE,EMAIL,SGRADE) values (%s,%s,%s,%s,%s) ON DUPLICATE KEY UPDATE SGRADE = VALUES(SGRADE)""" #to maintain atomicity
val=[
    (3,'manju',22,'manju@gmail.com',"7th semester"),
    (4,'puspa',20,'puspa@gmail.com',"2th semester"),
    (5,'rusan',23,'rusan@gmail.com',"8th semester"),
    (6,'Grishma',21,'grisma@gmail.com',"4th semester")
]
mycursor.executemany(sql,val)
mydb.commit()
print(mycursor.rowcount," records were inserted !")
```

```python
mycursor.execute("select * from tlb_students")
row=mycursor.fetchall()
for x in row:
    print(x)
```

```python
#removing duplicates based on name column
delete="""
DELETE t1 FROM tlb_students t1
INNER JOIN tlb_students t2 
WHERE t1.id > t2.id AND t1.NAME = t2.NAME;
"""
mycursor.execute(delete)
mydb.commit()
print(mycursor.rowcount,"duplicate rows deleted")

mycursor.execute("select * from tlb_students")
row=mycursor.fetchall()
for x in row:
    print(x)
```

```python
#using where, orderby and limit clause
import mysql.connector as sqlc
conn=sqlc.connect(
    host="localhost",
    user="root",
    password="",
    database="garimaDB"
)

#creating the cursor obbject
cursor=conn.cursor()
print("fetch data using different clause\n")
q="select * from tlb_students where id>=2 order by name asc limit 2"
try:
    cursor.execute(q)
    content=cursor.fetchall()
    for x in content:
        print(x)

except sqlc.Error as error:
    print("failed to seelct the record:", error)

finally:
    if conn.is_connected():
        cursor.close()
        conn.close()
        print("connection closed successfully !")
```
