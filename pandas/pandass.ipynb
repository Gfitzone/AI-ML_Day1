```python
#data cleaning
#removing empty cells
import pandas as pd
df=pd.read_csv('data.csv')
new_df=df.dropna() #by default dropna() returns a new DataFrame, and doesnot change the original
print(new_df.to_string())
```

```python
#to change we use inplace=True
import pandas as pd
df=pd.read_csv('data.csv')
df.dropna(inplace=True)
print(df.to_string())
```

```python
import pandas as pd
df=pd.read_csv('data.csv')
df.fillna({"Name":"Tom lee"},inplace=True)
df
```

```python
#replace empty cells with mean, median or mode
#calculate mean and replace any empty values with it

x=df["Age"].mean()
df.fillna({"Age":x},inplace=True)
```

```python
#calculate median and replace
import pandas as pd

df = pd.read_csv('data.csv')
x=df["Age"].median()
df.fillna({"Age":x},inplace=True)
print(df.to_string())
```

```python
#calculate mode and replace 
#mode is the value that appears most frequently
import pandas as pd
df=pd.read_csv('data.csv')
x=df["Age"].mode()[0]
df.fillna({"Age":x},inplace=True)
print(df.to_string())
```

```python
#cleaning the wrong data format
import pandas as pd
df=pd.read_csv('data.csv')
df['JoinDate']=pd.to_datetime(df['JoinDate'],errors='coerce')
#here errors=coerce will replace the invalid dates with NaT(not a time) which can be later filter or fill
print(df.to_string())
```

```python
#to remove those NaT we do
df.dropna(subset=['JoinDate'],inplace=True)
```

```python
print(df.to_string())
```

```python
#fixing wrong data
#setting email in row 6
import pandas as pd
df=pd.read_csv('data.csv')
df.loc[6,'Email']='mary@gmail.com'
print(df.to_string())
```

```python
#for larger datasets loop through all values 
#here in grade col, if the value is higher than 10 , set 10
for x in df.index:
    if df.loc[x,"grade"] > 11:
        df.loc[x,"grade"]=10

print(df.to_string())
```

```python
#we can also simply remove the row with the wrong data

df=pd.read_csv("data.csv")
for x in df.index:
    if df.loc[x,"grade"] > 11:
        df.drop(x, inplace=True)

```

```python
print(df.to_string())
```

```python
#removing duplicates rows
#duplicate rows are the rows that have been registered more than one time
#to find duplicate we use duplicated() method which returns boolean values for each row

import pandas as pd
df=pd.read_csv("data.csv")
print(df.duplicated())
#it will return True for every row that is duplicate otherwise False
```

```python
#to remove duplicate
df.drop_duplicates(inplace=True)
#inplace=True will make sure that the method doesnot return 
#the new dataframe instead it removes all duplicates from the original DataFrame
print(df.to_string()) 
```
