```python
#draw a line in a diagram from position (0,0) to position (6,250)

import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,6])
y=np.array([0,250])

plt.plot(x,y)
plt.show()
```

```python
# format strings fmt
#mark each point with a circle
y=np.array([3,2,5,3])
plt.plot(y,'o:r')
#o:r refers to marker|line|color which is the syntax of fmt
 #here r means red likewise 
#we have differ as g,b,c(cyan),m(magenta),y,k(black),w
plt.show()
```

```python
#line reference 
#we have -(solid line),:(dotted line),--(dashed), -.(dashed/dotted line)
y=np.array([4,5,8,9])
plt.plot(y,'3-.m') #syntax: marker|line|color
plt.show()
```

```python
#setting the size of the marker , we use markersize 'ms'
#and for the color of the marker edge , we use  marker-edge-color 'mec'
#for the color inside the edge marker , we use marker-face-color 'mfc'
y=np.array([4,6,7,9])
plt.plot(y,'o:c',ms=20,mec='r',mfc='b')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
#keyword argument linestyle or ls is used to change the style of the plotted line
u=np.array([11,2,3])
plt.plot(u,linestyle="dashdot") 
#we can also use symbols like : or ls in above
plt.show()
```

```python
#line color
u=np.array([3,2,5])
plt.plot(u,color='r') 
#we can use 'c' in place of color
#also can use hexadecimal color names as-> #4CAF50
plt.show()
```

```python
#line width (lw)
#plotting with 30.35pt width line
y=np.array([2,23,4,45])
plt.plot(y,linewidth='30.35')
plt.show()
```

```python
#multiple lines
#drawing two lines by specifying a plt.plot() function for each line
x=np.array([1,22,3,44])
y=np.array([5,66,7,88])
plt.plot(x)
plt.plot(y)
plt.show()
```

```python
#drawing two lines by specifying the 'x' and 'y' point values for both lines
x1=np.array([1,2,3,4])
x2=np.array([1,22,3,44])
#here x1 is x axis and x2 is y-axis

y1=np.array([4,5,6,7])
y2=np.array([6,77,8,99])
#here y1 is x-axis and y2 is y-axis

plt.plot(x1,x2)
plt.plot(y1,y2)

plt.show()
```

```python
#adding labels to both axis and adding a plot title
x=np.array([3,2,4,6])
y=np.array([200,100,400,700])

plt.plot(x,y)
plt.title("Demand and Supply diagram")
plt.xlabel("Demand")
plt.ylabel("Supply")
plt.show()
```

```python
#to set font for title and labels we can use fontdict parameter
x=np.array([2,3,4,6])
y=np.array([2,33,4,55])
plt.plot(x,y)

font1={'family':'serif','color':'magenta','size':20}
font2={'family':'Times New Roman','color':'green','size':11}

plt.title("Demand and supply chart",fontdict=font1,loc='left')
#loc means location of the title , it can be left, right and center(default)
plt.xlabel("Demand",fontdict=font2)
plt.ylabel("supply",fontdict=font2)

plt.grid() #this adds grid lines to the plot

plt.show()
```

```python
#we can also specify which grid lines to display
#legal values are 'x','y','both'(default)

x=np.array([2,3,4,5])
y=np.array([2,33,44,5])
plt.plot(x,y)
plt.grid(axis='x')
plt.show()
```

```python
#we can also set line properties of grid as
plt.plot(x,y)
plt.grid(color='magenta',linestyle='--',lw=0.5)
plt.show()
```

```python
#matplotlib subplot
#displaying multiple plot we use subplot() function

#plot1
x=np.array([2,3,4,5])
y=np.array([4,55,6,77])
plt.subplot(1,2,1)#it means it has one row , 2 colunm cause we are making 2 plot(if 3 then 3)and 1 means 1st plot
plt.plot(x,y)

#plot2
x = np.array([7, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(1,2,2)
plt.plot(x,y)

plt.show()
```

```python
#if i want multiple row and multiple column then i can do
#drawing 6 plots and give title to each

#plot1
x=np.array([2,3,4,5])
y=np.array([4,55,6,77])
plt.subplot(2,3,1)#it means it has 2 row , 3 colunm cause we are making 2 plot(if 3 then 3)and 1 means 1st plot
plt.plot(x,y)
plt.title("demand")

#plot2
x = np.array([7, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.title("supply")

#plot3
x=np.array([3,7,8,9])
y=np.array([55,6,88,7])
plt.subplot(2,3,3)
plt.plot(x,y)
plt.title("sales")


#plot4
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 4)
plt.plot(x,y)
plt.title("income")


#plot5
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])
plt.subplot(2, 3, 5)
plt.plot(x,y)
plt.title("profit")


#plot6
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])
plt.subplot(2, 3, 6)
plt.plot(x,y)
plt.title("loss")

#to add tiltle to the entire figure 
plt.suptitle("Garima-Thapa Company Limited")

plt.show()
```

```python
#creating scatter plot
x=np.array([2,3,4,5,4,5,6,7])
y=np.array([2,33,44,5,66,77,8,88])
plt.scatter(x,y)

x=np.array([2,4,4,7,1,2,3,4])
y=np.array([2,3,4,5,44,33,22,33])
plt.scatter(x,y,color="hotpink")
plt.show()
```

```python
#we can color each dot
#we use 'c' argument, color is not accepted

x=np.array([2,3,4,5])
y=np.array([33,22,55,10])
colors=np.array(['r','g','m','b'])
plt.scatter(x,y,c=colors)
plt.show()
```

```python
#color map is like a list of colors where 
# each color had a value from 0 to 100
#we use cmap to specify the colormap
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100]) 

plt.scatter(x,y,c=colors,cmap='viridis') 
# viridis is one of the built in colormaps in matplotlib
#there are many others as Accent,Blues ,Greens etc

plt.colorbar()#it will show the actual colormap
plt.show()
```

```python
#to change the size of the dots use 's' argument

x=np.array([3,4,2,6])
y=np.array([3,44,6,55])
size=np.array([20,63,144,12])
plt.scatter(x,y,s=size,alpha=0.5)
#here alpha is used to show the transparency of dots and is between 0 to 1
plt.show()
```

```python
#creating random arrays with 100 values for x-points
#y-points, colors and sizes

x=np.random.randint(100,size=(100))
y=np.random.randint(100,size=(100))
colors=np.random.randint(100,size=(100))
sizes=10*np.random.randint(100,size=(100))

plt.scatter(x,y,c=colors,s=sizes,alpha=0.5,cmap='tab10')
plt.colorbar()
plt.show()
```

```python
#creating bars, we use bar()
import numpy as np
import matplotlib.pyplot as plt
x=np.array(['a','b','c','d'])
y=np.array([63,40,6,90])
plt.bar(x,y,color="red",width=0.2)#here it will change 
#the size of width of the bar default is 0.8
plt.show()
```

```python
#to show bar horizontally we use barh()
x=['a','b','c','d']
y=[3,4,6,7]
plt.barh(x,y,height=0.4) #here in the horizontal bar 
#use height to change the size of the width
plt.show()
```

```python
#creating piecharts , we use pie()
import matplotlib.pyplot as plt
import numpy as np
a=np.array([34,22,13,17])
#to add labels use labels parameter
mylabels=["math","science","nepali","english"]
plt.pie(a,labels=mylabels)

plt.show()

```

```python
#starting first wedge at 180 degrees
plt.pie(a,labels=mylabels,startangle=180)
#default start angle is 0, there are 90,180 and 270 also 
#start angle is the starting point of the wedge(piece)
plt.show()
```

```python
#stand out the wedge, we use explode parameter
#pull the "math" wedge 0.2 from the center of the pie

myexplode=[0.2,0,0,0]
plt.pie(a,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

```python
#setting color to each wedge by myself
#we use colors parameter
mycolors=["magenta","cyan","k","#4CAF50"]
plt.pie(a,labels=mylabels,colors=mycolors)
#to add a list of explanation for each wedge
# use legend() and parameter 'title' if you want to add title
plt.legend(title='subjects')
plt.show()
```
