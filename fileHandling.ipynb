```python
f=open("demofile.txt")
print(f.read())
```

```python
#using with statement to open a file
with open("demofile.txt") as f:
    print(f.read())
    
```

```python
#closing a file...with statement automatically closes the file
#if you are not using with statement , you must close the file.
f=open("demofile.txt")
print(f.readline())
f.close()

#here the function of readline() is to read only one line of the file.
# If you want to read the whole file then you can use readlines() function.
#if you want to read 2 lines then you can use readline() function twice.
```

```python
#read only parts of the file
with open("demofile.txt") as f:
    print(f.read(5)) #this reads the first 5 characters of the file.
    
```

```python
#loop through the file line by line
with open("demofile.txt") as f:
    for x in f:
        print(x)
```

```python
#write to an existing file
#we must add a parameter to the open() function: a or w
with open("demofile.txt","a") as f:
    f.write("now the file has more content!")

#open and read the file after the appending:
with open("demofile.txt") as f:
    print(f.read())
```

```python
#overwiting the existing file
with open("myfile.txt","w") as f:
    f.write("now the file has beem overwriten")
    with open("myfile.txt") as f:
        print(f.read())
```

```python
#creating a new file we use either x ,w or a as the parameter to the open() function.
f=open("myfile.txt","x")
f=open("demofile.txt","x")
```
