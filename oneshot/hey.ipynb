```python
#define the function to calculate the area of a triangle
def calculate_triangle_area(base,height):
    area=0.5*base*height
    return area
#call the function with arguments
base=7
height=10
print(f"Area of a triangle with {base} and {height} is {calculate_triangle_area(base,height)}")

```

```python
#define a function to calculate the volume of a cylinder
def calculate_cylinder_volume(radius,height):
    pi=3.14
    volume=pi*radius**2*height
    return volume
radius=3
height=2
volumee=calculate_cylinder_volume(radius,height)   #volume mathi ko formula ma vako lekhda ni hunxa hai ,
#kei error aaaudaina tara maile yeha chai farak lekheko xu hai
print("The volume of a cylinder with radius ",radius," and height ",height, " is ",volumee)
```
