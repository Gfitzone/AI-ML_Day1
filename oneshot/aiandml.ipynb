```python
#create a function to print personal details of a person
def personal_details(name,age,address,qualification):
    print(f'Name:{name}')
    print(f'Age:{age}')
    print(f'Address:{address}')
    print(f'Qualification:{qualification}')
personal_details("sita",22,"urlabari","masters")
```

```python
#to avoid error while missing the argument we use argument which are of two types they are
#default and keyword arguments
#default argument
def personal_details(name="unknown name",age=0,address="unknown address",qualification="unknown"):
    print(f'name:{name}')
    print(f'Age:{age}')
    print(f'Address:{address}')
    print(f'Qualification:{qualification}')
personal_details()
```

```python
#to solve problem in default argu we use keyword where we use key value pair
personal_details(name="sita",address="urlabari")
```

```python
#error handling
#try, except(catch),else,finally
#try-> all runable code will be inside try block
#except->print message if error occurs
try:
    a=int(input("enter a number"))
    b=int(input("enter another number"))
    div=a/b
    print(div)
except ZeroDivisionError:
    print("Error: a number cannot be divisible by zero")
except ValueError:
    print("Error: only interger value is accepted")

```

```python
try:
    a=int(input("enter a number"))
    b=int(input("enter another number"))
    div=a/b
except ZeroDivisionError:
    print("Error: a number cannot be divisible by zero")
except ValueError: #to handle specific type of error
    print("Error: only interger value is accepted")
else: #all print value will be inside else block
    print(div)
finally: #finally block code will run either program runs or not
    print("program ended successfully")

```
