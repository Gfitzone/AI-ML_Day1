```python
#python program to check if the given number is prime or not using loop and
#break statement
n=int(input("Enter a number : "))
if n<2:
    print("not prime")
else:
    for i in range (2,n):
        if n%i==0:
            print(f"{n} is composite")
            #print(n," is not prime")
            break
    else:
        print(n," is prime")
```

```python
#program to find the largest element in a list using a loop and continue statement
#to skip the negative numbers
lst=list(map(int,input("enter a numbers separated by space: ").split()))
max_val=None
for num in lst:
    if num < 0:
        continue
    if max_val is None or num > max_val:
        max_val=num

print("largest elements from list:",lst ,"is" ,max_val)
```

```python
#calculate the sum of all even numbers between 1 and 100 
sum_even=0
for i in range(1,101):
    if i%2==0:
        sum_even += i
    else:
        pass #do nothing for odd numbers
print("sum of even numbers from 1 to 100 is ",sum_even)
```

```python
#program to print multiplication of a given numbers
n=int(input("enter a number"))
for i in range(1,11):
    #print(n,"x",i,"=",n*i)
    print(f"{n} X {i} = {n*i}")
```

```python
#to find the factorial of a given numbers
n=int(input("enter a number: "))
fact=1
for i in range(1,n+1):
    fact *= i

print("factorial of",n,"is",fact)
```

```python
#generate a fibonacci sequence up to 'n' terms using a loop
n=int(input("enter a number"))
a,b=0,1
for i in range(n):
    print(a,end=" ")
    a,b=b,a+b
```

```python
#to find the gcd of two numbers using loop and break
a=int(input("enter first number:"))
b=int(input("enter second number:"))
for i in range(min(a,b),0,-1):
    if a%i==0 and b%i==0:
        print(f"GCD of {a} and {b} is {i}")
       # print("GCD of",a,"and",b,"is",i)
        break #stop when found
```

```python
#check whether the given string is palindrome or not
#use loop and continue to ignore spaces
s=input("enter a string")
clean=""
for ch in s:
    if ch==" ":
        continue #skip spaces
    clean += ch

if clean==clean[::-1]:
    print(f"{s} is palindrome")
else:
    print(s,"is not palindrome")
```

```python
#print the ascii values of all the uppercase letters using a loop 
# and pass statement to handle lowercase letters

s=input("enter a string")
print("The entered string is ",s)
print("The ASCII values of only the uppercase letters is as follows:")
for ch in s:
    if ch.isupper():
        print(ch,":",ord(ch)) #prints ascii of uppercase

    else:
        pass #ignore lowercase
```

```python
#count the numbers of vowels and consonants in a string
s=input("Enter a string:").lower()
print("The entered string is:",s)
vowels=0
consonants=0
for ch in s:
    if ch.isalpha():
        if ch in 'aeiou':
            vowels += 1
        else:
            consonants +=1
print("vowels:",vowels)
print("consonants:",consonants)
```
