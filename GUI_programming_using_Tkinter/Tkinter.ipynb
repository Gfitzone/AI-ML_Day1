```python
#tkinter module is pythons standard GUI toolkit based on Tk
#used to create desktop applications with windows, buttons, menus and other graphical elements
'''
members/widgets of Tkinter

Button=widget for clickable buttons

Canvas=widget for drawing graphics

Entry= for single line text input

Frame=container widget for organizing other widget

Listbox=for displaying a list of items

Label=for displaying text or images

Menu=for creating menus

Text=for multi-line text input/display

Tk=class for creating the main application widget

AFTER CREATING THE WIDGET, TO DISPLAY IT WE NEED TO CALL A SPECIAL METHOD EITHER: PACK(),GRID() OR PLACE()
'''
```

```python
#check button widget -> to allow the user to select multiple options from a set of choices
from tkinter import *
master=Tk()
master.title("Check Buttons")
var1=IntVar()
Checkbutton(master,text="male",variable=var1).grid(row=0, sticky=W)
var2=IntVar()
Checkbutton(master,text="female",variable=var2).grid(row=1,sticky=W)

button=Button(master,text='save').grid(row=2,sticky=W,pady=4)
master.mainloop()
```

```python
#radio button widget
from tkinter import *
master=Tk()
master.title("Radio Button")

#for saving and exiting the window once the option is checked and save is clicked
def save_and_exit():
    if var.get()==1:
        print("male selected")
    elif var.get()==2:
        print("female selected")
    master.destroy()
    
var=IntVar()
Radiobutton(master,text="male",variable=var,value=1).grid(row=0,sticky=W) #sticky=W aligns the widget to the left edge of its cell.
Radiobutton(master,text="female",variable=var,value=2).grid(row=1,sticky=W)
button=Button(master,text='save',command=save_and_exit).grid(row=2,sticky=W,pady=4) #pady=4 means 4 pixels of extra space above and below the widget.
master.mainloop()
```

```python
#ListBox widget ->user can choose one or more alternatives from list of options
from tkinter import *
master=Tk()
master.title("List Box")

def save_and_exit():
    #get selected indices
    selected=lb.curselection()
    for i in selected:
        print(lb.get(i),"is selected")
    master.destroy()

lb=Listbox(master,selectmode=MULTIPLE) #allows multiple selections
lb.insert(1,"java")
lb.insert(2,"python")
lb.insert(3,"php")
lb.insert(4,"others")
lb.pack()

button=Button(master,text="save",command=save_and_exit).pack(pady=10)
mainloop()
```

```python
#scrollbar widget
from tkinter import *
root=Tk()
root.title("Scroll Bar")

#creating vertical scrollbar
scroll=Scrollbar(root)
scroll.pack(side=RIGHT,fill=Y) #stretches the widget vertically and place the scrollbar on the right side of the window


#Create a listbox widget and connect it to the scrollbar
# yscrollcommand=scroll.set means the scrollbar will update when the listbox scrolls
mylist=Listbox(root,yscrollcommand=scroll.set)
for line in range(20):
    mylist.insert(END,'This is line number'+ str(line))
    #here END means each new item is added to the end of the list

# Place the listbox on the left side and let it expand both horizontally and vertically
mylist.pack(side=LEFT,fill=BOTH)

# Link scrollbar and listbox
# Configure the scrollbar so it controls the vertical view of the listbox
scroll.config(command=mylist.yview)

root.mainloop()
```

```python
#menu widget
from tkinter import *
root=Tk()
root.title("Menu Example")
menu=Menu(root)
root.config(menu=menu)

filemenu=Menu(menu)
menu.add_cascade(label='file',menu=filemenu)
filemenu.add_command(label="new")
filemenu.add_command(label="open...")
filemenu.add_separator()
filemenu.add_command(label="exit",command=root.quit)

helpmenu=Menu(menu)
menu.add_cascade(label="help",menu=helpmenu)
helpmenu.add_command(label='about')

root.mainloop()
```

```python
#combobox widget->interactive dropdown menus that combine the feature of text fields and list boxes

import tkinter as tk
from tkinter import ttk

root=tk.Tk()
root.title("combobox example")
combo_box=ttk.Combobox(root,values=["option1","option2","option3"])
combo_box.pack(pady=4)

#set defalult value
combo_box.set("option1")

#bind event to selection
combo_box.bind("<<ComboboxSelected>>")
root.mainloop()
```
