```python
#find the GCD and LCM(least common multiple)

import math
a=int(input("enter 1st number"))
b=int(input("Enter 2nd number"))
gcd=math.gcd(a,b)
lcm=(a*b) // gcd
print("GCD of", a, "and", b, "is:", gcd)
print("LCM of", a, "and", b, "is:", lcm)
```

```python
#check the armstrong number
#armstrong number: sum of cubes of digits=number itself
#ex: 153

num=int(input("Enter a number"))
sum=0
temp=num

while temp>0:
    digit=temp%10
    sum += digit ** 3 # this means sum=sum+digit and digit=digit^3
    temp //= 10 # this means temp=temp/10

if num==sum:
    print(num,"is armstrong number")
else:
    print(num,"is not an armstrong number")
```

```python
#check the perfect numbers
#it means sum of divisors=number itself, eg:28-> 1+2+4+7+14=28

num=int(input("Enter a number"))
sum=0

for i in range(1,num):
    if num % i==0:
        sum += i

if sum==num:
    print(f"{num} is a perfect number")
else:
    print(f"{num} is not a perfect number...sum of divisors is {sum}")
```

```python
#sum of digits of a number
num=int(input("Enter a number"))
sum=0
temp=num
while num>0:
    digit=num%10
    sum += digit
    num //=10

print(f"sum of digits:({temp}) is {sum}")
```
