```python
#remove all punctuations from the string

import string

text="hello i am garima thapa! Are you happy?"

#syntax is 
# <expression> for <variable> in <iterable> if <condition>


clean_text="".join(ch for ch in text if ch not in string.punctuation)

#for ch in text , it loops through each character in text
# ch not in string checks and keeps only letters that are not punctuations
#join will join and "" means they are join with no separator
#here the ch at first means for each character that passes the filter, keep that character
#the first ch is value being returned and the second ch is loop variable , dont get confuse in that

print(f"original text: {text}")
print(f"without punctuations: {clean_text}")
```

```python
#find occurence of a substring

text="Dear unknown love, I am in love with you. may god bless you with love,care and prosperity "

substring="love"
count=text.count(substring)
print(f"({substring}) occurs {count} times in the text")

#user input
str=input("enter a string")
substr=input("enter a substring that you want to find the occurence of")
c=str.count(substr)
print(f"The entered text is: '{str}' and the occurence of {substr} is {c} times")
#print(f"({substr}) occurs {c} times in the text")
```

```python
#remove duplicates
str=input("enter a string")
result=""

for char in str:
    if char not in result:
        result += char

print("Original:", str)
print("Without duplicates:", result)
```
